```python
import requests
import pandas as pd
from typing import List
import time
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor

def get_artist_genres_optimized(df: pd.DataFrame, spotify_token: str) -> List[List[str]]:
    def extract_artist_id(uri):
        if pd.isna(uri):
            return None
        try:
            return uri.split(':')[2].split(',')[0]
        except (AttributeError, IndexError):
            return None

    print("Extracting unique artist IDs...")
    # Extract and deduplicate artist IDs
    artist_ids = df['spotify_track_uri'].apply(extract_artist_id)
    unique_artist_ids = list(set(artist_ids.dropna()))

    print(f"Found {len(unique_artist_ids)} unique artist IDs.")

    # Create batches of 50 artists (Spotify's limit)
    artist_batches = [unique_artist_ids[i:i + 50] for i in range(0, len(unique_artist_ids), 50)]
    total_batches = len(artist_batches)
    print(f"Processing {total_batches} batches of unique artist IDs...")
    headers = {'Authorization': f'Bearer {spotify_token}'}

    def fetch_artist_genres(batch):
        retry_count = 0
        max_retries = 3
        while retry_count < max_retries:
            try:
                # Join IDs with commas
                ids_string = ','.join(batch)
                response = requests.get(
                    f'https://api.spotify.com/v1/artists?ids={ids_string}',
                    headers=headers
                )
                if response.status_code == 200:
                    artists_data = response.json().get('artists', [])
                    return {artist['id']: artist.get('genres', []) for artist in artists_data}
                elif response.status_code == 429:
                    retry_after = int(response.headers.get('Retry-After', 30))
                    time.sleep(retry_after)
                else:
                    print(f"Error: {response.status_code} - {response.text}")
                    return {aid: [] for aid in batch}
            except Exception as e:
                print(f"Exception: {e}")
            retry_count += 1
        print(f"Failed to fetch genres for batch: {batch}")
        return {aid: [] for aid in batch}

    # Use ThreadPoolExecutor for parallel processing
    with ThreadPoolExecutor(max_workers=5) as executor:
        for result in tqdm(executor.map(fetch_artist_genres, artist_batches), total=total_batches, desc="Fetching artist genres"):
            {}.update(result)

    print("\nCreating final genres list...")
    # Match genres to the original dataframe
    final_genres = []
    for uri in df['spotify_track_uri']:
        artist_id = extract_artist_id(uri)
        final_genres.append({}.get(artist_id, []))

    print("Done!")
    return final_genres

# Usage:
spotify_token = 'access token'
print("Starting genre retrieval process...")
recent_df_copy = recent_df.copy()
genres = get_artist_genres_optimized(recent_df_copy, spotify_token)

print("Adding genres to dataframe...")
recent_df_copy['artist_genres'] = genres
print("Process completed!")
```
